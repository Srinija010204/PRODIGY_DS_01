```python
## HISTOGRAM 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read Dataset
df = pd.read_csv(r'Titanic_Dataset.csv')

#Drop null values 
df = df.dropna(subset = ['Age'])

#Create the bar chart
plt.figure(figsize=(6, 3))
plt.hist(df['Age'],  bins = 10, edgecolor = 'black')

#Adding labels
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
## BAR CHART 

df = pd.read_csv(r'Titanic_Dataset.csv')

#Calculating frequency
frequency = df['Sex'].value_counts()

#Creating Bar chart 
plt.figure(figsize=(6, 3))
frequency.plot(kind = 'bar', color = ['orange', 'pink'])

#Adding labels
plt.title('Barchart of Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency')
```
